**3.** Preprocessing

**3.1.** Scaling time and amount columns so they are in line with the other features [Transformation]

```python
from sklearn.preprocessing import RobustScaler

# Scaling amount and time feature values
df['amount_scaled'] = RobustScaler().fit_transform(df['Amount'].values.reshape(-1,1))
df['time_scaled'] = RobustScaler().fit_transform(df['Time'].values.reshape(-1,1))

# Removing unscaled amount and time columns
df.drop(['Time', 'Amount'], axis=1, inplace=True)
```

**3.2.** Create a subclass of non-fraudulent data points and undersample so there is a class balance [Cleaning]

```python
# Groups df by Class feature
grouped = df.groupby(df.Class)

# Creates 2 new dataframes of non-fraudulent and fraudulent data points
dfN = grouped.get_group(0)
dfF = grouped.get_group(1)

# Randomly undersampling the non-fraudulent datapoints
dfN = dfN.sample(492)

# Merge class-balanced dataframes into complete dataframe ready for training
merge = [dfN, dfF]
dfP = pd.concat(merge)

# Shuffle rows of dataframe to minimise variance so the model will generalise well to unseen data
dfS = dfP.sample(frac = 1, random_state = 53)

dfS.head()
```

```python
dfP.describe()
```

**3.3.** Split data into training and test data

```python
from sklearn.model_selection import train_test_split

X = dfS[['time_scaled', 'amount_scaled']]
y = dfS['Class']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

**4.** Test Classifier using Logistic Regression on time and amount features

```python
from sklearn.linear_model import LogisticRegression

logistic_regression = LogisticRegression()

logistic_regression.fit(X_train, y_train)
```

```python
y_pred = logistic_regression.predict(X_test)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
```

<h2>K-Nearest Neighbour Classification</h2>

```python
# Setting train/test data up
X = dfS[['V11', 'V14']]
y = dfS['Class']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

* Testing standard KNN

```python
from sklearn.neighbors import KNeighborsClassifier as knc

n = 5
knn_model = knc(n)

knn_model.fit(X_train, y_train)

y_pred = knn_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)

ConfusionMatrixDisplay.from_predictions(y_test, y_pred, cmap='Reds')
plt.title('V4 vs V14 K-NN, k = 5')

print("k = 5")
print("Precision: ", precision_score(y_test, y_pred))
print("Recall: ", recall_score(y_test, y_pred))
print("F1: ", f1_score(y_test, y_pred))
print("Accuracy: ", accuracy_score(y_test, y_pred))
```

* Finding best value for k

```python
from sklearn.model_selection import cross_val_score as cvs

k_values = [i for i in range (1,31)]
scores = []

for k in k_values:
    knn = knc(k)
    score = cvs(knn, X, y, cv=5)
    scores.append(np.mean(score))
```

```python
sns.lineplot(x = k_values, y = scores, marker = 'o')
plt.xlabel("K Values")
plt.ylabel("Accuracy Score")
plt.title("Finding optimum value for k (V11 vs V14)")
```

<h2>Perceptron</h2>

```python
from sklearn.linear_model import Perceptron

perc = Perceptron(max_iter=500, eta0=0.025)

X = dfS[['V11', 'V14']]
y = dfS['Class']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)

perc.fit(X_train, y_train)
```

```python
def plot_separation(X, y, perceptron):
    
    weights   = perceptron.coef_[0]
    bias      = perceptron.intercept_
    slope     = - weights[0]/weights[1]
    intercept = -bias/weights[1]
    
    xmin = min(dfS['V11'])
    xmax = max(dfS['V11'])
    ymin = min(dfS['V14'])
    ymax = max(dfS['V14'])

    x_line = np.linspace(xmin, xmax)
    y_line = slope * x_line + intercept
    
    fig = plt.figure(figsize=(8,6))
    
    plt.scatter(dfS['V11'], dfS['V14'], c=y, cmap='coolwarm', s=15)
    plt.plot(x_line, y_line, c='black', lw=4)
    
    plt.title("V11 vs V14 Perceptron Classifier - 200 iterations, 0.05 mutation")
    plt.xlabel("V11")
    plt.ylabel("V14")
    plt.xlim(xmin-1, xmax+1)
    plt.ylim(ymin-1, ymax+1)
    
    plt.show()
    
plot_separation(X, y, perc)
y_pred = perc.predict(X_test)
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_pred, cmap = 'Blues_r')
plt.title('V11 vs V14 Perceptron - 500 iterations, 0.025 mutation')

print("Precision: ", precision_score(y_test, y_pred))
print("Recall: ", recall_score(y_test, y_pred))
print("F1: ", f1_score(y_test, y_pred))
print("Accuracy: ", accuracy_score(y_test, y_pred))
```
